```python

df_positive_GDP = df[(df['GDP'] > 19098450000) & (df['GDP'] < 19098450000000)]


sns.scatterplot(data=df_positive_GDP, x = 'GDP', y = 'Life expectancy at birth (years)', 
                hue = 'Country')

```

```python
# First job: Get a scatter graph to view GDP vs Life Exp.

plt.figure(figsize = (10,10))
sns.set_style("darkgrid")
sns.color_palette("pastel")
#sns.scatterplot(data = df, x = 'Year', y = 'GDP', hue = 'Country')
sns.lmplot(data = df, x="Year", y="GDP", hue = 'Country')
plt.title('GDP over Time by Country')

```

```python
plt.figure(figsize = (10,10))
sns.set_style("darkgrid")
sns.color_palette("pastel")

df_lower_GDP = df[(df['Country'] != 'United States of America') & (df['Country'] != 'China')]


sns.lmplot(data = df_lower_GDP, x="Year", y="GDP", hue = 'Country')
plt.title('GDP over Time by Country (Excluding USA & China)')
```

```python
# Second job: Same graph but try to hue it by year, see whether relationship has changed.

plt.figure(figsize = (10,10))
sns.set_style("darkgrid")
sns.color_palette("pastel")
#sns.scatterplot(data = df, x = 'Year', y = 'GDP', hue = 'Country')
sns.lmplot(data = df, x="Year", y="Life expectancy at birth (years)", hue = 'Country');
plt.title('GDP over Time by Country')
```
